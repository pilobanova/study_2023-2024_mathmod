```julia
#  
using DifferentialEquations, Plots;

#    ,   
function f1(u, p, t)
    x, y = u
    a, b, c, d = p
    dx = a*x - b*x*y
    dy = -c*y + d*x*y
    return [dx, dy]
end

#  
u0 = [8,16]
p1 = [-0.83, -0.083, -0.82, -0.082]
tspan = (0.0,50.0)

#  
problem1 = ODEProblem(f1, u0, tspan, p1)

#   
sol1 = solve(problem1, Tsit5())

#  ,     
plot(sol1, title = " -", xaxis = "", yaxis = " ", label=["" ""])
```

```julia
plot(sol1, idxs=(1,2), xaxis="", yaxis="")
```
