```julia
# используемые библиотеки
using DifferentialEquations, Plots;

# задание системы дифференциальных уравнений, описывающих модель 
function f1(u, p, t)
    x, y = u
    a, b, c, d = p
    dx = a*x - b*x*y
    dy = -c*y + d*x*y
    return [dx, dy]
end

# начальные условия
u0 = [8,16]
p1 = [-0.83, -0.083, -0.82, -0.082]
tspan = (0.0,50.0)

# постановка проблемы
problem1 = ODEProblem(f1, u0, tspan, p1)

# решение системы ДУ
sol1 = solve(problem1, Tsit5())

# построение графика, который описывает изменение численности армий
plot(sol1, title = "Модель Лотки-Вольтерры", xaxis = "Время", yaxis = "Численость популяции", label=["жертвы" "хищники"])
```

```julia
plot(sol1, idxs=(1,2), xaxis="Жертвы", yaxis="Хищники")
```
