```julia
# используемые библиотеки
using DifferentialEquations, Plots;
function sir(u,p,t)
    (S,I,R) = u
    (a,b) = p
    N = S+I+R
    dS = -(a*S*I)/N
    dI = (a*I*S)/N - b*I
    dR = b*I
    return [dS, dI, dR]
end

N = 12400
I_0 = 150
R_0 = 55
S_0 = N - I_0 - R_0
u0 = [S_0, I_0, R_0]
p = [0.1, 0.05]
tspan = (0.0, 200.0)

prob = ODEProblem(sir, u0, tspan, p)
sol = solve(prob, Tsit5(), saveat = 0.1)
plot(sol, label = ["S" "I" "R"])
```
